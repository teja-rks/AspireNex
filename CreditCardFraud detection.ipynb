```python
# partition the data set
legit=df[df.Class==0]
```

```python
fraud=df[df.Class==1]
```

```python
legit['Class'].value_counts()
```

```python
fraud['Class'].value_counts()
```

```python
# making a sample dataset from the legit and the fraud dataset
legitsam=legit.sample(n=492)# why 492 because the fraud is 492 to make a perfect dataset we make equal of them
```

```python
#making the new dataset by concatinating the legitsam and fraud 
newdf=pd.concat([legitsam,fraud],axis=0)
```

```python
newdf
```

```python
newdf['Class'].value_counts()
```

```python
#spliting the data set into the feautres and the target 
x=newdf.drop(columns="Class",axis=1)
```

```python
y=newdf['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#importing the random forest classifier 
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
predicted=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(predicted,y_test)
```

```python
accuracy
```

```python
from sklearn.metrics import  classification_report
print(classification_report(y_test,predicted))
```
